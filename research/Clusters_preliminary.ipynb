#### 1. Well-Separated Clusters with Different Densities.


This dataset will test how well K-means and GMM can handle clusters with varying densities.

```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt

# Generate well-separated clusters with varying densities
centers = [[-10, -10], [0, 0], [10, 10], [-10, 10]]
cluster_std = [0.4, 1.5, 0.7, 0.3]  # Different densities

blobs_varying_density, _ = make_blobs(n_samples=1000, centers=centers, cluster_std=cluster_std, random_state=42)

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(blobs_varying_density[:, 0], blobs_varying_density[:, 1], s=10)
plt.title('Well-Separated Clusters with Varying Densities')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

#### 2. Non-Convex Shapes (Moons and Circles) with Noise


This dataset will challenge K-means and GMM while demonstrating the strength of DBSCAN and HDBSCAN.

```python
from sklearn.datasets import make_moons, make_circles

# Generate moons with noise
moons, _ = make_moons(n_samples=500, noise=0.1, random_state=42)

# Generate nested circles with noise
circles, _ = make_circles(n_samples=500, factor=0.5, noise=0.08, random_state=42)

# Combine datasets with an offset
circles[:, 0] += 2.5  # Shift circles to the right
complex_non_convex = np.vstack([moons, circles])

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(complex_non_convex[:, 0], complex_non_convex[:, 1], s=10)
plt.title('Non-Convex Shapes (Moons and Circles) with Noise')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

#### 3. Clusters with Outliers and Noise


This dataset will introduce outliers and noise to evaluate the robustness of DBSCAN and GMM.

```python
# Generate clusters with outliers
core_samples, _ = make_blobs(n_samples=500, centers=[[0, 0], [5, 5], [-5, -5]], cluster_std=0.5, random_state=42)
outliers = np.random.uniform(low=-10, high=10, size=(50, 2))  # 50 random outliers

# Combine the core samples and outliers
data_with_outliers = np.vstack([core_samples, outliers])

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(data_with_outliers[:, 0], data_with_outliers[:, 1], s=10)
plt.title('Clusters with Outliers and Noise')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

#### 4. Clusters with Varying Shapes and Densities


This dataset combines different shapes, sizes, and densities, ideal for testing all algorithms under complex conditions.

```python
# Generate a variety of clusters
blobs1, _ = make_blobs(n_samples=300, centers=[[0, 0]], cluster_std=0.2, random_state=42)
blobs2, _ = make_blobs(n_samples=300, centers=[[5, 5]], cluster_std=1.5, random_state=42)
blobs3, _ = make_blobs(n_samples=300, centers=[[-5, 5]], cluster_std=0.5, random_state=42)
moons, _ = make_moons(n_samples=300, noise=0.1, random_state=42)
circles, _ = make_circles(n_samples=300, factor=0.3, noise=0.05, random_state=42)

# Combine all different shapes
complex_shapes = np.vstack([blobs1, blobs2, blobs3, moons, circles])

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(complex_shapes[:, 0], complex_shapes[:, 1], s=10)
plt.title('Clusters with Varying Shapes and Densities')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

#### 5. Overlapping Clusters with Noise


This dataset includes overlapping clusters and noise, challenging for all algorithms, especially K-means and Agglomerative.

```python
# Generate overlapping clusters
overlap_centers = [[0, 0], [1.5, 1.5], [3, 3]]
overlapping_clusters, _ = make_blobs(n_samples=1000, centers=overlap_centers, cluster_std=1.0, random_state=42)

# Add noise
noise = np.random.uniform(low=-3, high=6, size=(100, 2))  # 100 random noise points
overlapping_with_noise = np.vstack([overlapping_clusters, noise])

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(overlapping_with_noise[:, 0], overlapping_with_noise[:, 1], s=10)
plt.title('Overlapping Clusters with Noise')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

#### 6. Hierarchical/Nested Structures


This dataset will test the performance of hierarchical clustering algorithms.

```python
from sklearn.datasets import make_blobs

# Create nested blobs
nested_blobs1, _ = make_blobs(n_samples=200, centers=[[0, 0]], cluster_std=0.1, random_state=42)
nested_blobs2, _ = make_blobs(n_samples=200, centers=[[0, 0]], cluster_std=0.3, random_state=42)
nested_blobs3, _ = make_blobs(n_samples=200, centers=[[0, 0]], cluster_std=0.5, random_state=42)

# Combine to create nested structure
nested_structure = np.vstack([nested_blobs1, nested_blobs2, nested_blobs3])

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(nested_structure[:, 0], nested_structure[:, 1], s=10)
plt.title('Hierarchical/Nested Structures')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
